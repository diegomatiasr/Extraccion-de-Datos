**ARCHIVOS FILTRADOS A CSV**

```python
# A. LOS REGISTROS DE 5 "HOST NAME" DIFERENTES
nombres = sicilia_data.iloc[0:6,4]  
print(nombres)

nombres = sicilia_data[(sicilia_data['host_name'] == 'Yumi Adriano') | (sicilia_data['host_name'] == 'Giuseppe') | (sicilia_data['host_name'] == 'Serena') | (sicilia_data['host_name'] == 'Gaspare') | (sicilia_data['host_name'] == 'Francesco')] 
nombres
nombres.to_csv("HostName.csv")
```

```python
#ConversiÃ³n de tipo de dato de columna de tipo Object a Float
sicilia_data['host_since']= sicilia_data['host_since'].astype('datetime64[s]')
```

```python
# B. LOS REGISTROS DE LOS HOST QUE SE UNIERON A AIRBNB "DESPUÃ‰S DEL AÃ‘O 2020"
since_2020 = sicilia_data[(sicilia_data["host_since"] >= "01/01/20")]
since_2020

since_2020.to_csv("Since2020.csv")
```

```python
# C. LOS REGISTROS DE LOS HOST QUE RESPONDEN "A MÃS TARDAR EN 1 DÃA"
tiempo_respuesta = sicilia_data[(sicilia_data["host_response_time"] == 'within a day') | (sicilia_data["host_response_time"] == 'within an hour') | (sicilia_data["host_response_time"] == 'within a few hours')]
tiempo_respuesta

tiempo_respuesta.to_csv("ResponseTime.csv")
```
