```python
# Archivo CSV de mi ciudad
sicilia_data= pd.read_csv("Sicily.csv").head()
pd.options.display.max_rows = 200
sicilia_data
```

```python
sicilia_data.info()
```

```python
# Ajustar maximo de columnas
pd.options.display.max_columns = None
```

```python
#Eliminar un signo de una columna 
sicilia_data['host_response_rate']=sicilia_data['host_response_rate'].str.replace('%', '')
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
sicilia_data['host_response_rate']= sicilia_data['host_response_rate'].astype(float)
```

```python
# a) Los registros de los host cuyo "host_response_rate" sea mayor a 95% y que sean "superhost"
filtroa = sicilia_data[(sicilia_data["host_response_rate"] >= 95) & (sicilia_data["host_is_superhost"] == "t")]
filtroa  

#filtroa.to_csv("ResponseRate-SuperHost.csv")
```

```python
# b) Los registros cuyas propiedades no tengan "1 baño"
filtrob = sicilia_data[(sicilia_data["bathrooms"] != 1.0)]
filtrob

#filtroa.to_csv("Baño.csv")
```

```python
# c) Los registros que tengan un precio entre "$100 y $200 US" (Realizar conversión de moneda)
filtroc = sicilia_data[(sicilia_data["price"] >= "$100") & (sicilia_data["price"] <= "$200")]
filtroc  

#filtroc.to_csv("Entre100-200-SuperHost.csv")
```

```python
# d) Los registros que sean "host_is_superhost" y que hayan verificado su perfil "host_identity_verified"
filtrod = sicilia_data[(sicilia_data["host_is_superhost"] >= 't') & (sicilia_data["host_identity_verified"] == "t")]
filtrod

#filtrod.to_csv("SuperHost-IdentityVerified.csv")
```

```python
# e) Los registros de los tipos de cuarto: "hotel room" o "Private room"
filtroe = sicilia_data[(sicilia_data["room_type"] == "Hotel room") | (sicilia_data["room_type"] == "Private room")]
filtroe

#filtroe.to_csv("HotelRoom-PrivateRoom.csv")
```

```python
# f) Los registros que tengan más de 4 "accommodates"
filtrof = sicilia_data[(sicilia_data["accommodates"] > 4)]
filtrof

#filtrof.to_csv("Accommodates.csv")
```
